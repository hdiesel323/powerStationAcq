```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the geodatabase
gdf = gpd.read_file('Substations.gdb')

# Display the first few rows
print(gdf.head())

# Plot the substations
gdf.plot()
plt.title("Electric Power Substations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
gdf.plot(marker='o', color='blue', markersize=5, figsize=(10, 10))
plt.title("Location of Electric Power Substations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
gdf.plot(column='MAX_VOLT', cmap='viridis', legend=True, figsize=(10, 10))
plt.title("Electric Power Substations by Max Voltage")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
